# K-Means

### Standardize First

#### 0: Youngest, 100% Female, Mostly Single, Not Worried About Savings - Seemingly
#### 1: Long Hotel Stays, Highest Savings, Around Average Age For Segments (41), Second Highest Chance Of Dependents,  Highest Hotel Savings, Highest Flight Booking Proportion
#### 2: Second Youngest, All Male, Mostly Single - Not Worried About Savings - Seemingly
#### 3: Oldest Group, Mostly Married With Children, Equivalent Sex Ratio, Most Worried About Savings
#### 4: Longest Hotel Stays, 75% Women, Most Likely Single Moms or smaller likelihood of Married w/o Children, Book Flights Alot and are only a little concerned with Hotel Savings, Middle Aged

##### 0: Millenial Single Women ======================================================================== Free Checked Bag
##### 1: Middle Aged Travelers Who Stay At Hotels Alot (Backpackers or Business People) ======================== Exclusive Discounts
##### 2: Millenial Single Men ========================================================================== No Cancellation Fees
##### 3: Older Married People - Probaly Seasonal Trips / See College Aged Children ================================== Free Hotel Meal
##### 4: People Staying In Hotels The Most, Probably Business Trips, Savings aren't eminently important ===== 1 Night Free Hotel With Flight

```python
perk_order = ["Free Checked Bag", 
           "Exclusive Discounts", 
           "No Cancellation Fees", 
           "Free Hotel Meal", 
           "1 Night Free Hotel With Flight"]
perk_map = dict(zip(range(len(perk_order)), perk_order))
perk_map
```

```python
def workflow(DF_data, perk_map, num_clusters):
    raw_unscaled_data = DF_data.copy()
    
#     GET THE SCALED DATA USING SCALER 
    raw_scaled_data = get_scaled_data(raw_unscaled_data)
    
#     RUN THE K-MEANS ALGORITHM
    kmeans = KMeans(n_clusters=num_clusters, random_state=0, n_init='auto')
    kmeans.fit(raw_scaled_data)
    
#     COPY ORIGINAL DATA FOR USE LATER
    unscaled_segmented_data = raw_unscaled_data.copy()
    scaled_segmented_data = raw_scaled_data.copy()
    
    
#     ADD THE K-MEANS LABELS FOR FURTHER PROCESSING
    unscaled_segmented_data["Segment"] = kmeans.labels_
    scaled_segmented_data["Segment"] = kmeans.labels_
    
    
#     CREATE NEW DFs FOR GETTING THE AVERAGES OF ACCUMULATED DATA THUS FAR TO DETERMINE WHAT LABELS TO APPLY TO EACH
    unscaled_data_means = unscaled_segmented_data.groupby(["Segment"]).mean()
    scaled_data_means = scaled_segmented_data.groupby(["Segment"]).mean()
    
    
    
#     REPLACE THE INDEX OF THE AGGREGATED DATA TO BE THE PERKS INSTEAD
    scaled_data_means = scaled_data_means.replace(perk_map)

    
#     CREATE THE LABEL COLUMN TO ASSOCIATE PERKS AND LABELS RECEIVED
    unscaled_segmented_data["Segment"] = unscaled_segmented_data["Segment"].astype("int64")
    unscaled_segmented_data["Label"] = unscaled_segmented_data["Segment"].replace(perk_map)
    
    scaled_segmented_data["Segment"] = scaled_segmented_data["Segment"].astype("int64")
    scaled_segmented_data["Label"] = scaled_segmented_data["Segment"].replace(perk_map)
    
    
#     k_std, k_data = update_means_and_data(k_std, k_data, perk_map)
   
    
    
#     RETURN DATA FOR USE IN FURTHER ITERATIONS
    return (raw_unscaled_data, 
            raw_scaled_data,
            unscaled_segmented_data,
            scaled_segmented_data,
            unscaled_data_means,
            scaled_data_means)
```

```python
main_data = procure_all_data(group)

starting_data = main_data.drop(columns=["age_group", "user_id"])
num_clusters = 5

raw_unscaled_data, \
raw_scaled_data, \
unscaled_segmented_data, \
scaled_segmented_data, \
unscaled_data_means, \
scaled_data_means = workflow(
   starting_data,
    perk_map, 
    num_clusters
 )

# #     USE PCA TO SEE HOW VARIABLES RELATE
pca = get_pca(raw_scaled_data)
pca.explained_variance_ratio_

num_components = len(starting_data.columns)
components = [f"Component {x+1}" for x in range(num_components)]
component_df = get_component_df(starting_data, pca, components)

if run_plots:
    variance_chart(pca)
    compare_hmap(component_df, components)
##     REITERATE TO GET A BETTER UNDERSTANDING OF VARIABLES NEEDED
```
